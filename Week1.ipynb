## Simple call to OpenAPI for code generation call

```python
from dotenv import load_dotenv
import os
load_dotenv()

gemini_key = os.getenv("GEMINI_API_KEY")
```

```python
from openai import OpenAI
def write_code(prompt):
    userMessage = [{"role":"user","content":prompt}]
    messages = [{"role":"system", "content":"You are software engineer who can write c# code."}]
    messages.extend(userMessage)
    print(messages)
    openAI = OpenAI(api_key=gemini_key,base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
    response = openAI.chat.completions.create(
        model="gemini-2.5-flash", messages=messages)
    return response.choices[0].message.content
```

```python
code = write_code("Give Highly efficient code to check if the given string is palindrome")
print(code)
```
